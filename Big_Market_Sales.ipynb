Imputing Item_weight

```python
#Determining average weight per item
item_avg_weight=data.pivot_table(values='Item_Weight',index='Item_Identifier')
#variable specifiying missing rows
bool_w=data['Item_Weight'].isnull()
#printing no of rows having null values before imputing
print('no of rows before imputing')
print(sum(bool_w))
#imputing missing values in Item_weight
data.loc[bool_w,'Item_Weight']=data.loc[bool_w,'Item_Identifier'].apply(lambda x: item_avg_weight[x])
#printing no of rows for checking of null values after imputing
print('no of rows of item_weight after imputing')
print(sum(data['Item_Weight'].isnull()))



```

Imputing Output_size

```python
#importing mode
from scipy.stats import mode
outlet_size_mode=data.pivot_table(values='Outlet_Size',columns='Outlet_Type',aggfunc=(lambda x: mode(x).mode[0]))
print('mode for each outlet')
print(outlet_size_mode)
#Maintaining bool variable for seeing where Outlet_Size have null values
bool_outlet_size=data['Outlet_Size'].isnull()
#printing null values before imputing
print('Before imputing no of null Output_Size')
print(sum(bool_outlet_size))
#imputing Output_Size
data.loc[bool_outlet_size,'Outlet_Size']=data.loc[bool_outlet_size,'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
#checking if any null value is present or not
print('After imputing no of null values')
print(sum(data['Outlet_Size'].isnull()))
```

```python
#define taregtet and IDcol
target='Item_Outlet_Sales'
IDcol=['Item_Identifier','Outlet_Identifier']
from sklearn import cross_validation,metrics
def modelfit(alg,dtrain,dtest,predictors,target,IDcol,filename):
    #Fit the algorithm on data
    alg.fit(dtrain[predictors],dtrain[target])
    #Predict Training set
    dtrain_predictions=alg.predict(dtrain[predictors])
     #Perform cross-validation:
    cv_score = cross_validation.cross_val_score(alg, dtrain[predictors], dtrain[target], cv=20, scoring='mean_squared_error')
    cv_score = np.sqrt(np.abs(cv_score))
    
    #Print model report:
    print "\nModel Report"
    print "RMSE : %.4g" % np.sqrt(metrics.mean_squared_error(dtrain[target].values, dtrain_predictions))
    print "CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score))
    #predictions of test data
    dtest[target]=alg.predict(dtest[predictors])
    
     #Export submission file:
    IDcol.append(target)
    submission = pd.DataFrame({ x: dtest[x] for x in IDcol})
    submission.to_csv(filename, index=False)
```
